# Data Formatting

```python
#removing the underscore from 'electric bike' and 'classic bike' from the 'rideable_type' column
divvy_data['rideable_type'] = divvy_data['rideable_type'].str.replace(r'_', ' ')
```

```python
#capitalizing the first letters from the 'rideable_type' and 'member_casual' columns
divvy_data['rideable_type'] = divvy_data['rideable_type'].str.title()
divvy_data['member_casual'] = divvy_data['member_casual'].str.title()
```

```python
#updating 'started_at/ended_at' columns to be datetime data type
divvy_data[['started_at', 'ended_at']] = divvy_data[['started_at', 'ended_at']].apply(pd.to_datetime)
```

```python
#separating out the datetime values from the 'started_at/ended_at' columns to individual date columns
divvy_data['start_date'] = divvy_data['started_at'].dt.date
divvy_data['end_date'] = divvy_data['ended_at'].dt.date
```

```python
#updating 'started_at_date/ended_at_date' columns to be date data type
divvy_data[['start_date', 'end_date']] = divvy_data[['start_date', 'end_date']].apply(pd.to_datetime)
```

```python
#adding a column for the trip lenth in seconds
divvy_data['TripLength'] = (divvy_data['ended_at'] - divvy_data['started_at']).dt.total_seconds()
```

```python
#removing the '*' from the "end_station_name" column
#divvy_data['end_station_name'] = divvy_data['end_station_name'].map(lambda x: x.strip('*'))
divvy_data['end_station_name'] = divvy_data['end_station_name'].astype(str).str.strip('*')
```

```python
divvy_data.sort_values(by='TripLength', ascending=True)
```

```python
#renaming the column names in each dataframe
divvy_data = divvy_data.rename(columns={'ride_id':'RideID', 'rideable_type':'RideableType', 'started_at':'StartedAt', 'ended_at':'EndedAt', 'start_station_name':'StartStationName', 'start_station_id':'StartStationID', 'end_station_name':'EndStationName', 'end_station_id':'EndStationID', 'start_lat':'StartLat', 'start_lng':'StartLng', 'end_lat':'EndLat', 'end_lng':'EndLng', 'member_casual':'MemberCasual', 'start_date':'StartDate', 'end_date':'EndDate'})
```

```python
#adding a column to account for the hour for 'Started At Time/Ended At Time' for grouping purposes
divvy_data['HourStarted'] = pd.to_datetime(divvy_data['StartedAt'], format='%H:%M:%S').dt.hour
```

```python
#adding a column to account for the month for 'Start Date/End Date' for grouping purposes
divvy_data['StartMonth'] = pd.DatetimeIndex(divvy_data['StartDate']).month

#changing month number to month name
divvy_data['TripMonth'] = divvy_data['StartMonth'].apply(lambda x: calendar.month_name[x])
```

```python
#adding day of week column for trips taken 
divvy_data['DayofWeek'] = divvy_data['StartDate'].dt.day_name()
```

```python
#removing rows that have a negative value in 'TripLength'
trip_length_negative = (divvy_data[divvy_data['TripLength'] <= 0].index)
trip_length_negative
```

```python
#removing rows that are less than 60 seconds in 'TripLength'
trip_length_60 = divvy_data[divvy_data['TripLength'] <= 61].index
trip_length_60
```
